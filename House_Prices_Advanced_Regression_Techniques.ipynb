#LinearRegression ,Polynomial Features ve Ridge islenmesi

```python
from sklearn.preprocessing import StandardScaler,FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.preprocessing import PolynomialFeatures
```

```python
log_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('trans',FunctionTransformer(np.log,feature_names_out='one-to-one')),
    ('scale',StandardScaler())
])


cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features),
    ('log', log_pipeline, ['LotFrontage'])
],remainder='passthrough')

estimator=LinearRegression()

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('poly',PolynomialFeatures(degree=2,include_bias=False)),
    ('ridge',Ridge(alpha=0.1))
    # ('estimator',estimator)
])


full_pipeline.fit( X_train,y_train)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

#RandomForestRegressor and GridSearch

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
log_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('trans',FunctionTransformer(np.log,feature_names_out='one-to-one')),
    ('scale',StandardScaler())
])


cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features),
    ('log', log_pipeline, ['LotFrontage'])
],remainder='passthrough')

estimator=RandomForestRegressor()

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('estimator',estimator)
])


full_pipeline.fit( X_train,y_train)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
params={
    'estimator__n_estimators':[100,120,140,160],
    'estimator__max_depth':[10,15,20,25]
}

grid_search1=GridSearchCV(full_pipeline, params, n_jobs=-1)

grid_search1.fit(X_train, y_train)
```

```python
grid_search1.best_params_
```

```python
random_forest=grid_search1.best_estimator_
random_forest.fit(X_train,y_train)
random_forest.score(X_train,y_train),random_forest.score(X_test,y_test)
```

# GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

estimator=GradientBoostingRegressor(learning_rate=0.05,min_samples_leaf=10)

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('estimator',estimator)
])


full_pipeline.fit( X_train,y_train)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(full_pipeline, X_train, y_train, cv=5, scoring='r2')
print(f"Cross-validation scores: {cv_scores}")
```

#Voting

```python
from sklearn.ensemble import VotingRegressor
```

```python
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

voting=VotingRegressor(
    estimators=[
        ('lnr',LinearRegression()),
        ('rfr',RandomForestRegressor(random_state=42)),
        ('gbr',GradientBoostingRegressor(random_state=42))
        ]
)

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('voting',voting)
])


full_pipeline.fit( X_train,y_train)
```

```python
for name,clf in voting.named_estimators_.items():
  print(name,full_pipeline.score(X_test,y_test))
```

```python
full_pipeline.score(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

#Stacking

```python
from sklearn.ensemble import StackingRegressor
```

```python
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

stacking=StackingRegressor(
    estimators=[
        ('lnr',LinearRegression()),
        ('rfr',RandomForestRegressor(random_state=42)),
        ('gbr',GradientBoostingRegressor(random_state=42))
        ],
    final_estimator=RandomForestRegressor(random_state=42)
)

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('stacking',stacking)
])


full_pipeline.fit( X_train,y_train)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```
