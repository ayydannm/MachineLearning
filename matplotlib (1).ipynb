```python
data={
    'year':[2019,2020,2021,2022,2023,2024,2025],
    'population':[9981500,10023000,10027000,10050000,10100000,10180770,10423252]
}
```

```python
import pandas as pd
df=pd.DataFrame(data)
plt.figure(figsize=(10,6))
plt.gca().set_facecolor('#baf7f7') #gca()  "get current axes"
plt.grid(True,linestyle='--',color='black',alpha=0.5)
plt.plot(df['year'],df['population'],color='blue',linewidth=3,marker='o',markersize=10,markerfacecolor='maroon')
plt.title("Population of country(2019-2025)",fontsize=16,fontweight='bold',color='darkslategray')
plt.xlabel('Year',fontsize=12,fontweight='bold',color='darkslategray')
plt.ylabel('Population',fontsize=16,fontweight='bold',color='darkslategray')

for x,y in zip(df['year'],df['population']):
  plt.text(x,y+6000,f"{y:,}",ha='right',fontsize=10,fontweight='bold',color='black')

plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
import pandas as pd
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease['target'].value_counts().plot(kind='bar');
```

```python
heart_disease.hist(figsize=(10,10));
```

```python
heart_disease['age'].hist(bins=20)
```

```python
heart_disease.plot.scatter(x = 'age', y = 'chol');
```

```python
import matplotlib.pyplot as plt
colors = ['red' if t == 1 else 'blue' for t in heart_disease['target']]

plt.scatter(
    heart_disease['age'],
    heart_disease['chol'],
    c=colors
);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=heart_disease['age'],y=heart_disease['chol'],c=heart_disease['target'])
ax.set(title='Heart disease and cholesterol levels',xlabel='Age',ylabel='Cholestrol');
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(
    x = heart_disease['age'],
    y = heart_disease['chol'],
    c = heart_disease['target'] #targetdeki deyerlere gore reng ver
)

plt.colorbar(scatter)
ax.set(title='Heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol');
```

```python
fig, ax = plt.subplots(figsize=(10,6))

colors = ['red' if t == 1 else 'blue' for t in heart_disease['target']]

ax.scatter(
    heart_disease['age'],
    heart_disease['chol'],
    c=colors
)

ax.set(
    title='Heart disease and cholesterol levels',
    xlabel='Age',
    ylabel='Cholesterol'
)

plt.show()
```

```python
plt.figure(figsize=(6,5))
heart_disease.boxplot(column='age', by='target')
plt.title('Age vs Heart Disease')
plt.suptitle('')
plt.xlabel('Target (0=No, 1=Yes)')
plt.ylabel('Age')
plt.show();
```
