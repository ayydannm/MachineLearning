```python
X=df.drop('medv',axis=1)
y=df['medv'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
import numpy as np
num_features=X_train.select_dtypes(include=[np.number]).columns
```

```python
num_features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.preprocessing import PolynomialFeatures
```

```python
num_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features)
],remainder='passthrough')

estimator=LinearRegression()

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('estimator',estimator)

])
```

```python
full_pipeline.fit(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
from sklearn.metrics import  mean_squared_error
import numpy as np

y_pred = full_pipeline.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))

rmse
```

#PolynimialFeatures+LinearRegression

```python
poly_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('estimator', LinearRegression())
])
```

```python
poly_pipeline.fit(X_train, y_train)
```

```python
poly_pipeline.score(X_train, y_train)
```

```python
poly_pipeline.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# 5-fold CV
cv_scores = cross_val_score(poly_pipeline, X_train, y_train, cv=5, scoring='r2')

print("CV R2 Scores:", cv_scores)
print("Mean CV R2 Score:", np.mean(cv_scores))
```

#Ridge regularization

```python
poly_ridge_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('estimator', Ridge(alpha=1))
])
```

```python
poly_ridge_pipeline.fit(X_train, y_train)
```

```python
train_score = poly_ridge_pipeline.score(X_train, y_train)
test_score = poly_ridge_pipeline.score(X_test, y_test)
print("Train R2 Score:", train_score)
print("Test R2 Score:", test_score)
```

```python
cv_scores = cross_val_score(poly_ridge_pipeline, X_train, y_train, cv=5, scoring='r2')
print("CV R2 Scores:", cv_scores)
print("Mean CV R2 Score:", np.mean(cv_scores))
```

#Lasso,ElasticNet

```python
lasso_pipeline = Pipeline([
    ('scaler', transformer),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('estimator', Lasso(alpha=0.1))
])

# ElasticNet pipeline
en_pipeline = Pipeline([
    ('scaler', transformer),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('estimator', ElasticNet(alpha=0.1, l1_ratio=0.5))
])

# Fit və skorlar
for name, pipeline in zip(['Lasso', 'ElasticNet'], [lasso_pipeline, en_pipeline]):
    pipeline.fit(X_train, y_train)
    train_score = pipeline.score(X_train, y_train)
    test_score = pipeline.score(X_test, y_test)
    cv_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='r2')
    print(f"\n{name} Model")
    print("Train R2:", train_score)
    print("Test R2:", test_score)
    print("CV R2:", np.mean(cv_scores))
```

#RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('estimator',RandomForestRegressor(n_estimators=100, random_state=42))
])
```

```python
rf_pipeline.fit(X_train, y_train)
```

```python
train_score = rf_pipeline.score(X_train, y_train)
test_score = rf_pipeline.score(X_test, y_test)
print("Train R2:", train_score)
print("Test R2:", test_score)
```

```python
cv_scores = cross_val_score(rf_pipeline, X_train, y_train, cv=5, scoring='r2')
print("CV R2 Scores:", cv_scores)
print("Mean CV R2:", np.mean(cv_scores))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 0.5, 'sqrt']
}

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("Best params:", grid_search.best_params_)
print("Best CV R2:", grid_search.best_score_)

# Ən yaxşı modeli testdə yoxla
best_rf = grid_search.best_estimator_
print("Test R2:", best_rf.score(X_test, y_test))
```

```python
# Feature-lər DataFrame-dən
feature_names = X.columns

# Feature importance
importances = best_rf.feature_importances_

# DataFrame ilə sırala
feat_imp_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(feat_imp_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(feat_imp_df['Feature'], feat_imp_df['Importance'], color='skyblue')
plt.gca().invert_yaxis()  # ən əhəmiyyətli feature yuxarda
plt.xlabel('Importance')
plt.title('Random Forest Feature Importance')
plt.show()
```
