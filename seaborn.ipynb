```python
cars=sns.load_dataset('mpg')
cars
```

```python
type(cars)
```

```python
cars.shape
```

```python
cars.isnull().sum()
```

```python
cars.dropna(inplace=True)
cars.shape
```

```python
sns.relplot(x='model_year',y='mpg',col='origin',hue='cylinders',data=cars);
```

```python
sns.kdeplot(cars.horsepower,fill=True); #Kernel Density Estimation plot
```

```python
sns.kdeplot(
    x=cars.horsepower,
    y=cars.mpg,
    levels=15,      # neÃ§É™ kontur xÉ™tti
    fill=True,
);
```

```python
sns.histplot(cars.horsepower,kde=True,stat='density')
```

```python
sns.histplot(x='horsepower',data=cars,hue='origin')
```

```python
sns.histplot(x='model_year',hue='origin',multiple='dodge',data=cars,shrink=0.9)
```

```python
cars.columns
```

```python
sns.pairplot(data=cars, vars=["model_year", "horsepower", "weight"],hue='origin')
```

```python
sns.boxplot(x='origin', y='mpg', data=cars)
```

```python
sns.violinplot(x='origin', y='mpg', data=cars, palette='Set2');
```

```python
sns.jointplot(x='horsepower', y='mpg', data=cars,kind='reg');
```

```python
corr = cars.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
