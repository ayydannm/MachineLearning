```python
if df['price_x'].isna().sum()>0:
  df=df.dropna(subset=['price_x'],inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df["price_x"])
plt.show()
```

```python
Q1 = df["price_x"].quantile(0.25)
Q3 = df["price_x"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df["price_x"] >= lower_bound) & (df["price_x"] <= upper_bound)]
```

```python
df['currency_x'].value_counts()
```

```python
currency_rates = {'$': 1.7, '€': 1.85, 'AZN': 1}
df['price_x'] = df.apply(lambda row: row['price_x'] * currency_rates.get(row['currency_x'], 1), axis=1)
```

```python
df.drop(columns=['currency_x'],inplace=True)
```

```python
df['Yerlərin sayı']=df['Yerlərin sayı'].str.replace('[/+]','',regex=True)
df['Yürüş']=df['Yürüş'].str.replace('[/km]','',regex=True)
df['Ban növü'] = df['Ban növü'].str.split(r'[/,]').str[0]
```

```python
df[['Həcmi', 'At Gücü', 'Yanacaq Növü']] = df['Mühərrik'].str.split(r'[/,]', expand=True)
df['At Gücü'] = df['At Gücü'].str.replace(' a.g.', '')
df['Yanacaq Növü'] = df['Yanacaq Növü'].str.strip()
```

```python
df.drop(columns=['Mühərrik'],inplace=True)
```

```python
df['Yerlərin sayı']=pd.to_numeric(df['Yerlərin sayı'],errors='coerce')
df['Yürüş']=pd.to_numeric(df['Yürüş'],errors='coerce')
df['At Gücü']=pd.to_numeric(df['At Gücü'],errors='coerce')
```

```python
df['extra_info'].value_counts().head(20)
```

```python
features = ['Yüngül lehimli disklər', 'ABS', 'Lyuk', 'Yağış sensoru', 'Mərkəzi qapanma',
            'Park radarı', 'Kondisioner', 'Oturacaqların isidilməsi', 'Dəri salon',
            'Ksenon lampalar', 'Arxa görüntü kamerası', 'Yan pərdələr', 'Oturacaqların ventilyasiyası']

for feature in features:
    df[feature] = df['extra_info'].str.contains(feature, na=False).astype(int)
```

```python
df.head()
```

```python
df.drop(columns=['extra_info'],inplace=True)
```

```python
df['Vəziyyəti'] = df['Vəziyyəti'].str.replace(',', '', regex=True)
```

```python
X=df.drop('price_x',axis=1)
y=df['price_x'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
cat_features=X_train.select_dtypes(exclude=[np.number]).columns
num_features=X_train.select_dtypes(include=[np.number]).columns
```

#LinearRegression

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.preprocessing import PolynomialFeatures
```

```python
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

estimator=LinearRegression()

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('ridge',Ridge(alpha=0.1))
    # ('estimator',estimator)
])


full_pipeline.fit(X_train[:5000],y_train[:5000])
```

```python
full_pipeline.score(X_train[:5000],y_train[:5000])
```

```python
full_pipeline.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = full_pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mse
```

#RandomForestRegressor and GridSearchCV

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

estimator=RandomForestRegressor()

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('estimator',estimator)
])


full_pipeline.fit(X_train[:5000],y_train[:5000])
```

```python
full_pipeline.score(X_train[:5000],y_train[:5000])
```

```python
full_pipeline.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(full_pipeline, X_train[:5000], y_train[:5000], cv=4, scoring='r2')
```

```python
print("Cross-validation R^2 Nəticələri: ", cv_scores)
print("Ortalama R^2: ", cv_scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV
params={
    'estimator__n_estimators':[100,120,140],
    'estimator__max_depth':[10,15,20]
}

grid_search1=GridSearchCV(full_pipeline, params, cv=3, n_jobs=-1)

grid_search1.fit(X_train[:5000], y_train[:5000])
```

```python
grid_search1.best_params_
```

```python
random_forest=grid_search1.best_estimator_
random_forest.fit(X_train[:5000],y_train[:5000])
random_forest.score(X_train[:5000],y_train[:5000]),random_forest.score(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
feature_importances = full_pipeline.named_steps['estimator'].feature_importances_
feature_names = transformer.get_feature_names_out()

sorted_idx = np.argsort(feature_importances)[::-1]
plt.figure(figsize=(10,6))
plt.barh(np.array(feature_names)[sorted_idx][:10], feature_importances[sorted_idx][:10])
plt.xlabel("Importance Score")
plt.ylabel("Feature Name")
plt.title("Top 10 Important Features")
plt.gca().invert_yaxis()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(full_pipeline, X_train[:5000], y_train[:5000],
                            cv=5, scoring='r2', n_jobs=-1)

print("Cross-validation R^2 Nəticələri: ", cv_scores)
print("Ortalama R^2: ", cv_scores.mean())
```

```python
from sklearn.metrics import mean_squared_error

y_pred = full_pipeline.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print(f"MAE: {mae}")
```

#Voting

```python
from sklearn.ensemble import VotingRegressor
```

```python
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

voting=VotingRegressor(
    estimators=[
        ('lnr',LinearRegression()),
        ('rfr',RandomForestRegressor(random_state=42)),
        ('gbr',GradientBoostingRegressor(random_state=42))
        ]
)

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('voting',voting)
])


full_pipeline.fit(X_train[:5000],y_train[:5000])
```

```python
full_pipeline.score(X_train[:5000],y_train[:5000])
```

```python
full_pipeline.score(X_test,y_test)
```
